```python
from glob import glob
import pandas as pd
import os
import shutil
from skimage.io import imread,imsave
import numpy as np
```

```python
plates = []
for i in range(5,17):
    plates.append('P%sa-20230508_wHaloCaMP' %str(i).zfill(2))
imgPath = r'Z:\GECIScreenData\GECI_Imaging_Data\wHaloCaMP'
date = '20230522_wHaloCaMP_raw'
savePath = r'Y:\GECI_screening\wHaloCaMP\combo & single mutants 2023'+'\\'+date
if not os.path.exists(savePath):
    os.makedirs(savePath)
print(os.path.exists(savePath))
plates
```

```python
for plate in plates:
    if not os.path.exists(os.path.join(imgPath,date,plate,'imaging')):
        os.mkdir(os.path.join(imgPath,date,plate,'imaging'))
    stimFiles = glob(os.path.join(imgPath,date,plate) + '\\*sweep*')
    refFiles = glob(os.path.join(imgPath,date,plate) + '\\*ref*')
    no_stim_files = glob(os.path.join(imgPath,date,plate) +'\\*noStim*')
    stimFiles = sorted(stimFiles)
    refFiles = sorted(refFiles)
    no_stim_files = sorted(no_stim_files)
    df_well = pd.DataFrame()
    wells = []
    for stimFile in stimFiles:
        wells.append(os.path.basename(stimFile).split('_')[1])
    df_well = df_well.assign(well=wells)
    for well in df_well.well.unique():
        wellPath = os.path.join(imgPath,date,plate,'imaging',well)
        if os.path.exists(wellPath):
            continue
        os.mkdir(wellPath)
        for stimFile in stimFiles:
            if os.path.basename(stimFile).split('_')[1] == well:
                shutil.move(stimFile,os.path.join(imgPath,date,plate,'imaging',well,os.path.basename(stimFile)))
        os.mkdir(os.path.join(imgPath,date,plate,'imaging',well,'Ref_images'))
        for refFile in refFiles:
            if os.path.basename(refFile).split('_')[1] == well:
                shutil.move(refFile,os.path.join(imgPath,date,plate,'imaging',well,'Ref_images',os.path.basename(refFile)))
        os.mkdir(os.path.join(imgPath,date,plate,'imaging',well,'noStim'))
        for  no_stim_file in no_stim_files:
            if os.path.basename(no_stim_file).split('_')[1] == well:
                shutil.move(no_stim_file,os.path.join(imgPath,date,plate,'imaging',well,'noStim'),os.path.basename(no_stim_file))
```
